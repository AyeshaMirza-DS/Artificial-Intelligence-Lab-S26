BFS with Queue 

```python
from collections import deque

# graph
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

visited = []
queue = deque()

start = 'A'

visited.append(start)
queue.append(start)

while queue:
    node = queue.popleft()
    print(node)

    for neighbour in graph[node]:
        if neighbour not in visited:
            visited.append(neighbour)
            queue.append(neighbour)
```

BFS without Queue

```python
# Very simple BFS without queue

graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

visited = []  # keeps track of visited nodes

def bfs(node):
    if node not in visited:
        print(node)
        visited.append(node)
        for neighbour in graph[node]:
            bfs(neighbour)

# start BFS from A
bfs('A')
```
