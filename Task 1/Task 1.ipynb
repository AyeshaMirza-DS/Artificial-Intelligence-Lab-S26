Dynamic To-Do List (Using List & Loop)

```python
tasks = []                                                        # Creat an empty list

while True:                                                      # Creat an infinit loop
    print("\n---TO DO LIST ---")                                 # Print the title
    print("1. Add Task ")
    print("2. View Task ")
    print("3. Remove Task ")
    print("4. Exit ")

    choice= input(" Enter your choice (1-4):")                  #Take input from user

    #Add Task
    if choice=="1":                                             # Apply Condition
        task = input(" Enter new task:")
        tasks.append(task)                                       # add new task into the list
        print(f"Task , {task},  added successfully! ")          # Print confirmation msg

    #View Task
    elif choice=="2":
        if len(tasks) == 0:                                      # Check kry ga aghr list empty hai
            print("No task available. ") 
        else:                                                   #Condition, if list is not empty
            print(" Your Task ")
            for i, task in enumerate(tasks, start=1):            # i --> task number
                print(f" {i}.{task}")                           # Print task nbr & name
    # Remove Task
    elif choice == "3":
        if len(tasks) == 0:
            print("No task to remove.")
        else:
            for i, task in enumerate(tasks, start=1):
                print(f"{i}. {task}")

            num= int(input("Enter task number to remove: "))

            if 1 <= num <= len(tasks):                                        #checks if nbr is valid  
                removed = tasks.pop(num - 1)                                  # pop --> for removing task
                print(f"Task , {removed}, removed successfully! ")
            else:                                                            #Condition, if wrong nbr entered
                print(" Invalid task number. ")

    #Exit
    elif choice == "4":
        print("Goodbye!")
        break                                                   # Stop the loop 

    else:                                                       #aghr user 1-4 k elava koi nbr enter kry ga  
        print("Invalid choice. Please try again.")              #Shows error msg  

```
