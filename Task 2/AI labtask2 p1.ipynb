MiniProject 1: 
Today's project is actually a very common interview question, which revolves around a childhood 
counting game called Fizz Buzz. 

```python
previous = 0
for i in range(1, 101):
    new_num = i+previous

    if new_num % 3 == 0 and new_num % 5 == 0:
        print("Fizz Buzz")
    elif new_num % 3 == 0:
        print("Fizz")
    elif new_num % 5 == 0:
        print("Buzz")
    else:
        print(new_num)

    previous = i
```
