### Checking null values

```python
Dataset.isnull().sum()
```

## Filling null values

```python
num = Dataset['company'].mode()[0]
Dataset['company'] = Dataset['company'].fillna(num)
```

```python
num = Dataset['agent'].mode()[0]
Dataset['agent'] = Dataset['agent'].fillna(num)
```

```python
num = Dataset['country'].mode()[0]
Dataset['country'] = Dataset['country'].fillna(num)
```

```python
Dataset.isnull().sum()
```

### Dropping unnecessary columns

```python
Dataset.drop('required_car_parking_spaces', axis=1, inplace=True)
```
