```python
# 1. Numeric columns (children, agent, company) ko Median se fill kia hai
Dataset = Dataset.fillna(Dataset.median(numeric_only=True))

# 2. Text columns (country) ko Mode se fill kia hai
Dataset['country'] = Dataset['country'].fillna(Dataset['country'].mode()[0])

Dataset.isnull().sum()
```
